## Phase 3 — Feature Engineering & Cleaning

### 3.1 — Extract Date Features

```python
df['displacement_start_date'] = pd.to_datetime(df['displacement_start_date'])
df['displacement_end_date']   = pd.to_datetime(df['displacement_end_date'])

df['duration_days'] = (df['displacement_end_date'] - df['displacement_start_date']).dt.days
df['month']         = df['displacement_date'].dt.month
df['year']          = df['displacement_date'].dt.year

print("Duration stats:")
print(df['duration_days'].describe().round(2))
```

### 3.2 — Extract Region from Location Name

```python
# Extract the first region mentioned in locations_name
# Format is usually "District, Region, Somalia"
def extract_region(loc):
    if pd.isna(loc):
        return 'Unknown'
    parts = loc.split(',')
    # Region is usually the second part
    if len(parts) >= 2:
        return parts[1].strip()
    return parts[0].strip()

df['region'] = df['locations_name'].apply(extract_region)

# Clean up region names — keep top regions, group rare ones as Other
top_regions = df['region'].value_counts().head(10).index.tolist()
df['region'] = df['region'].apply(lambda x: x if x in top_regions else 'Other')

print("Top regions:")
print(df['region'].value_counts())
```

### 3.3 — Engineer Target Column

```python
df['is_large_event'] = (df['figure'] > 100).astype(int)
print("Target distribution:")
print(df['is_large_event'].value_counts())
print(df['is_large_event'].value_counts(normalize=True).round(3))
```

### 3.4 — Select Final Features

```python
FEATURES = [
    'displacement_type',   # Conflict or Disaster
    'combined_type',       # Conflict, Drought, Flood
    'month',               # month of event
    'duration_days',       # how long the event lasted
    'latitude',            # geographic location
    'longitude',
    'year',                # 2025 or 2026
    'region',              # extracted region name
]
TARGET = 'is_large_event'

df_clean = df[FEATURES + [TARGET]].copy()
print("Clean dataset shape:", df_clean.shape)
df_clean.head()
```

### 3.5 — Handle Missing Values

```python
print("Missing values:")
print(df_clean.isnull().sum())
print()

# duration_days: fill with median
df_clean['duration_days'] = df_clean['duration_days'].fillna(df_clean['duration_days'].median())

# latitude/longitude: fill with Somalia centroid
df_clean['latitude']  = df_clean['latitude'].fillna(5.152)
df_clean['longitude'] = df_clean['longitude'].fillna(46.199)

print("After imputation:")
print(df_clean.isnull().sum())
```

### 3.6 — IQR Cap on duration_days

```python
def iqr_cap(series, k=1.5):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    return series.clip(lower=q1 - k * iqr, upper=q3 + k * iqr)

df_clean['duration_days'] = iqr_cap(df_clean['duration_days'])
df_clean['duration_days'].describe().round(2)
```
